```python
import pandas as pd
```

```python
toronto_profile_csvpath = "Resources/neighbourhood-profiles-2016-csv.csv"

toronto_2016_raw_df = pd.read_csv(toronto_profile_csvpath)
toronto_2016_raw_df.head(5)
```

```python
df = toronto_2016_raw_df
df.columns
```

```python
df = df.drop(columns=['_id', 'Category','Data Source','Characteristic','Topic','City of Toronto','Topic'])
df
```

```python
dfT = df.T
dfTi = dfT.iloc[:,0:3]
dfTi
```

```python
dfT2 = dfT.iloc[:,9:15]
dfT2
```

```python
dft_merge = dfTi.merge(dfT2,how='outer',left_index=True,right_index=True)
dft_merge
```

```python
dft_merge = dft_merge.drop(columns=[1])

dft_merge
```

```python
dft_merge.columns =['Neighbourhood ID','Population','Children (0-14 years)','Youth (15-24 years)','Working Age (25-54 years)','Pre-retirement (55-64 years)','Seniors (65+ years)','Older Seniors (85+ years)'] 
dft_merge.reset_index()
dft_merge.rename(columns={"index":"Neighbourhood"})
dft_merge
```
